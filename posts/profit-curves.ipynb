##Profit Curves

- Combining **prediction** made by model **and business requirements**
- Calculate **Expected Profit** based on how much a **correct prediction makes us** and how much **a false one costs us**
- Based on where the **Maximum Expected Profit (MEP)** is, we can then **set the threshold for predicting +ve**
- Other constraints such as how much the budget is, so we can only do **MEP given budget limits**

##Case Study: Predicting Churn Rate

- **Correctly predicting churn** helps us do **targeted customer retention schemes**
- **+ X**  by correctly predicting churn (**True positive**)
- **- Y**  by incorrectly predicting churn (**False positive**) (Offer non-churn cust. sweet deals)
- **- 0** by correctly predicting non-churn behavior (**True Negative**)
- **- 0** by incorrectly predicting non-churn behavior (**False Negative**)
    - Would seem like we lose something by no detecting churn when in fact it is 
    - But since **TP is related to FN, ie more TP, less FN**, we **don't double count FN**
 
<br>
```
Cost Benefit Matrix

|TP FP|  |+X -Y|
|TN FN|  |-0 -0|  
```

````python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression as LR
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn
%matplotlib inline
````

####Calculate the rate give a sklearn generated confusion matrix 

- Output: [TPR, FPR], [FNR, TNR])

````python
def confusion_rates(cm):
    [[tn, fp], [fn, tp]] = cm

    N = fp + tn
    P = tp + fn

    tpr = tp / P
    fpr = fp / N
    fnr = fn / P
    tnr = tn / N

    return np.array([[tpr, fpr], [fnr, tnr]])
````

**Predict churn by setting different thresholds for detecting positive**

- Starting from the highest threshold for getting positive
- Gradually moving down the threshold to predict more churners
- Increasing both TP and FP as we go
- But there is a sweet spot where we can get the most TP with lowest FP given the cost-benefit matrix

````python
def profit_curve(classifiers, cb, X_train, y_train, X_test, y_test, axes):
    #Percentage of positive
    pos = np.sum(y_test == 1) / len(y_test)
    neg = 1 - pos
    class_probs = np.array([pos, neg])
    
    for classifier in classifiers:
        # SVM needs to have probabilities turned on to use them
        if classifier.__name__ == 'SVC':
            model = classifier(probability=True)
        else:
            model = classifier()
        model.fit(X_train, y_train)
        probabilities = model.predict_proba(X_test)[:,1]
        #reverse sort index (from large to small)
        indicies = np.argsort(probabilities)[::-1]

        profit = []
        for i in xrange(len(indicies)):
            pred_false = indicies[i:]
            y_predict = np.ones(len(indicies))
            #Starting from the largest threshold for predicting positive
            #Everything started -ve with max threshold (p) for positive
            #Each loop move one p down
            #Could be a problem if there are a lot with same probability
            y_predict[pred_false] = 0
            rates = confusion_rates(confusion_matrix(y_test, y_predict))
            profit.append(np.sum(class_probs * rates * cb))

        percentages = np.arange(len(indicies)) / len(indicies) * 100
        axes.plot(percentages, profit, label=classifier.__name__);

    axes.legend(loc="lower left");
    axes.set_title("Profits of classifiers");
    axes.set_xlabel("Percentage of test instances (decreasing by score)");
    axes.set_ylabel("Profit");
    #plt.ylim(20)
````

````python
def prepare_data(filename):
    churn_df = pd.read_csv(filename)

    # Clean up categorical columns
    churn_df['Churn?'] = (churn_df['Churn?'] == 'True.').astype(int)
    yes_no_cols = ["Int'l Plan", "VMail Plan"]
    churn_df[yes_no_cols] = (churn_df[yes_no_cols] == "yes").astype(int)

    # set label array
    y = churn_df.pop('Churn?').values

    # Drop unwanted columns
    churn_df = churn_df.drop(['State', 'Area Code', 'Phone'], axis=1)

    # set feature matrix
    X = churn_df.values

    # Train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y)

    # Scale data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    return X_train, y_train, X_test, y_test
````

````python
X_train, y_train, X_test, y_test = prepare_data('data/churn.csv')

fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(2, 2, figsize=(12,12))

# Cost-Benefit Matrix
cb1 = np.array([[79, -1],
               [0, 0]])

cb2 = np.array([[79, -10],
               [0, 0]])

cb3 = np.array([[79, -20],
               [0, 0]])

cb4 = np.array([[79, -50],
               [0, 0]])


# Define classifiers for comparison
classifiers = [RF, LR, GBC, SVC]

# Plot profit curves
profit_curve(classifiers, cb1, X_train, y_train, X_test, y_test, ax1)
profit_curve(classifiers, cb2, X_train, y_train, X_test, y_test, ax2)
profit_curve(classifiers, cb3, X_train, y_train, X_test, y_test, ax3)
profit_curve(classifiers, cb4, X_train, y_train, X_test, y_test, ax4)

````

As we increase the penalty (**-1 --> -50**) for **falsely predicting churn**

- Expected profit decreases
- Maximum Expected Profit is reduced and reached much quicker
- Meaning we have to be extra careful about who we give the free goods to 
- Since they cost a lot (Relative to the return if indeed they were about to churn)
- In the -1 case, we can simply predict everyone as churn and give them stuff
- Since the cost is low for FP
