##Support Vector Classifier
- One of the most **accurate** (On par with Neural Nets)
- Takes a **long time to train** (Much longer then Random Forest / Logit)
- Takes even **longer to CrossValidate**
- Sensitive to hyperparameters (Much tuning required - GridSearchCV)
- Still used in industry when the accurate is needed
- Continously training the SVM with a moving window of data
- Partial_fit might be able to speed things up 
- Averaging of parameters on diff. subset of data can help too (Not in Python)
- Data should be **scaled (z-score, Normal Standard)**, where in Random Forest, doesn't really matter

##Definition of HyperPlane / Class Boundary

Flat surface of dimension $p - 1$, where $p$ is number of features

$\beta_0 + \beta_1x + \beta_2x = 0$

##Outcome for SVC

The outcome we are trying to predict is binary $(-1, 1)$

i.e.

$\beta_0 + \beta_1x + \beta_2x > 0$ then predicted value is $1$

$\beta_0 + \beta_1x + \beta_2x < 0$ then predicted value is $-1$

##Three Types of SVC

1. Maximal Margin Classifier
1. Support Vector Classifer
1. Support Vector Machine

##Difference from Logistic Regression
- Logistic Regression is a specification of SVC
- Which opitimizes for the best performance in prediction based on training set
- **Low Bias, High Variance**, ie relatively small change in one feature can cause a change in class
- **SVC** emphasizes on maximum margin which increase the amount of change in features to cause a change in class
- **SVC only takes into account the points that are close to the hyperplane / class boundary for classification**
- **Logit takes all the data points into account (more suspectible to outliers)**


##Maximal Margin Classifier

- **Suport Vectors** are the points **closest from the Hyperplane**
- **Margin:** Distance between **Support vectors** and **Hyperplane** (_**see diagram**_)
- The objective is to maximize margin 
- The **larger the margin, the less chance of misclassification**, because there is more separation between the groups
- There is a **margin function** to judge how confident we are about the classification:

$margin = y^{(i)} (w^T x)$,

Positive Margin = Correct prediction. How positive indicates how good the prediction is

$y^{(i)} = -1$ and $(w^T x) < 0$, positive margin 

$y^{(i)} = 1$ and $(w^T x) > 0$, positive margin

##Support Vector Classifier

- When classes are **not linearly separable**
- Must tweak the **C parameter** to specify the **balance between minimizing error and maximizing margin**
- C is inverse $\lambda$ (in regularization), where **higher C = Lower bias, Higher Variance**
- **Increase C** if you want **less training error, but also less margin** (which could mean more test error)

##Support Vector Machine

- When the classes are not linearly separable 
- **> 2 classes**
- A **linear kernel is insufficient** for the problem at hand
- Need another kernel, e.g. **rbf (Radial Basis Function)**, **poly (Polynomial)**
- A more complex function to describe the hyperplane



###Radial Basis Function

```python
parameters = {'C':[0.0001,0.005, 0.1,10, 100, 1000], 'gamma':10.0 ** np.arange(-5, 4)}
est = SVC(kernel="rbf")
GridSearchCV(est, parameters, cv=3)
```

###Polynomial Kernel

```python
parameters = {'C':[0.0001,0.005, 0.1,10, 100, 1000], 'degree': [0,1,2,3]}
est = SVC(kernel="poly")
GridSearchCV(est, parameters, cv=3)
```

In summary, few things you can tweak with SVC

- **Continuous**
    - C (regularization term)
    - L1 / L2 loss function (type of regularization)
    - class_weight (sample ratio)
    

- **Binary-Class**
    - C (regularization term)
    - class_weight (sample ratio)


- **Multi-Class / Non-linear**
    - C (regularization term)
    - class_weight (sample ratio)
    - kernel (rbf mostly) 
    - kernel dependent params (gamma)
